```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
p_1 = pd.read_csv('obj_p.1_cv.dat', sep = '  ', delimiter = 'Objective function: ', header=None, engine='python') 
data_1 = np.zeros((len(p_1),4))
for i in range(len(p_1)):
    data_1[i] = p_1[1][i].split()
    
data_1 = pd.DataFrame(data_1, columns= ["MSE Test","Act Term","Wt Bs Term","Obj Funct"])
data_1 = data_1.drop_duplicates().sort_values(by="Obj Funct")[0:500]
data_1 = data_1.reset_index()
data_1["Complx"] = data_1["Act Term"]+data_1["Wt Bs Term"]
data_1["RMSE Test [km/s]"] = np.sqrt(data_1["MSE Test"])
data_1
```
